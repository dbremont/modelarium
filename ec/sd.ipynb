```python
import numpy as np
import matplotlib.pyplot as plt
```

## Model

```python
# Price range (x-axis for supply and demand curves)
prices = np.linspace(0, 50, 100)

# Demand function: Quantity demanded as a function of price
def demand(p):
    return 100 - 2 * p  # Example demand curve: Q = 100 - 2P

# Supply function: Quantity supplied as a function of price
def supply(p):
    return 2 * p  # Example supply curve: Q = 2P

# Calculate quantity demanded and supplied for each price
quantity_demanded = demand(prices)
quantity_supplied = supply(prices)

# Equilibrium price and quantity (where demand equals supply)
equilibrium_price = 25  # Set equilibrium price
equilibrium_quantity = demand(equilibrium_price)  # Calculate quantity at equilibrium price
```

```python
# Adjusting the axes so that Quantity is on the x-axis and Price is on the y-axis

# Plotting the supply and demand curves with adjusted axes
plt.figure(figsize=(8, 6))
plt.plot(quantity_demanded, prices, label="Demand Curve", color='blue')
plt.plot(quantity_supplied, prices, label="Supply Curve", color='red')

# Mark the equilibrium point (quantity, price)
plt.plot(equilibrium_quantity, equilibrium_price, 'go', label=f'Equilibrium Point ({equilibrium_quantity}, {equilibrium_price})')

# Labels and title
plt.title("Supply and Demand Curves")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
